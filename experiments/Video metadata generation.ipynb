````python
from youtube_transcript_api import YouTubeTranscriptApi
import requests


def get_video_chapters(video_id):
    api_url = f"https://yt.lemnoslife.com/videos?part=chapters&id={video_id}"
    response = requests.get(api_url)

    if response.status_code == 200:
        data = response.json()
        return data["items"][0]["chapters"]["chapters"]
    else:
        print(f"Failed to retrieve chapters. Status code: {response.status_code}")
        return None


def get_video_transcript(video_id):
    return YouTubeTranscriptApi.get_transcript(video_id)


def combine_chapters_and_captions(chapters, transcript):
    combined_list = []

    for i, chapter in enumerate(chapters):
        chapter_title = chapter["title"]
        chapter_start = chapter["time"]
        chapter_end = (
            chapters[i + 1]["time"]
            if i + 1 < len(chapters)
            else transcript[-1]["start"] + transcript[-1]["duration"]
        )

        # Combine captions within this chapter's time range
        chapter_captions = " ".join(
            caption["text"]
            for caption in transcript
            if chapter_start <= caption["start"] < chapter_end
        )

        combined_list.append(
            {
                "title": chapter_title,
                "start": chapter_start,
                "end": chapter_end,
                "captions": chapter_captions,
            }
        )

    return combined_list


# Example usage
video_id = "yBBdUlZ-h4U"
chapters = get_video_chapters(video_id)
transcript = get_video_transcript(video_id)

if chapters and transcript:
    result = combine_chapters_and_captions(chapters, transcript)
    for item in result:
        print(item)
````

````python
import requests
from youtube_transcript_api import YouTubeTranscriptApi


def get_video_details(video_id, youtube_api_key):
    # Function to get video metadata
    def get_video_metadata(video_id):
        api_url = f"https://www.googleapis.com/youtube/v3/videos?part=snippet%2Cstatistics&id={video_id}&key={youtube_api_key}"
        response = requests.get(api_url)

        if response.status_code == 200:
            data = response.json()
            if data["items"]:
                item = data["items"][0]
                title = item["snippet"]["title"]
                author = item["snippet"]["channelTitle"]
                date_of_sharing = item["snippet"]["publishedAt"]
                likes_count = item["statistics"].get("likeCount", "N/A")
                return title, author, date_of_sharing, likes_count
            else:
                raise ValueError(
                    "No video information found for the provided video ID."
                )
        else:
            raise Exception(
                f"Failed to retrieve video metadata. Status code: {response.status_code}"
            )

    # Function to get video chapters
    def get_video_chapters(video_id):
        api_url = f"https://yt.lemnoslife.com/videos?part=chapters&id={video_id}"
        response = requests.get(api_url)

        if response.status_code == 200:
            data = response.json()
            return data["items"][0]["chapters"]["chapters"]
        else:
            raise Exception(
                f"Failed to retrieve chapter data. Status code: {response.status_code}"
            )

    # Function to get video transcript
    def get_video_transcript(video_id):
        return YouTubeTranscriptApi.get_transcript(video_id)

    # Function to combine chapters and captions
    def combine_chapters_and_captions(chapters, transcript):
        combined_list = []

        for i, chapter in enumerate(chapters):
            chapter_title = chapter["title"]
            chapter_start = chapter["time"]
            chapter_end = (
                chapters[i + 1]["time"]
                if i + 1 < len(chapters)
                else transcript[-1]["start"] + transcript[-1]["duration"]
            )

            # Combine captions within this chapter's time range
            chapter_captions = " ".join(
                caption["text"]
                for caption in transcript
                if chapter_start <= caption["start"] < chapter_end
            )

            combined_list.append(
                {
                    "chapter_title": chapter_title,
                    "chapter_start": chapter_start,
                    "chapter_end": chapter_end,
                    "chapter_captions": chapter_captions,
                }
            )

        return combined_list

    # Get video metadata
    video_title, video_author, video_date, video_likes = get_video_metadata(video_id)

    # Get chapters and transcript
    chapters = get_video_chapters(video_id)
    transcript = get_video_transcript(video_id)

    # Combine chapters and captions
    chapters_and_captions = combine_chapters_and_captions(chapters, transcript)

    # Combine the metadata and chapter/caption data
    final_output = {
        "video_title": video_title,
        "video_author": video_author,
        "video_date_of_sharing": video_date,
        "video_likes_count": video_likes,
        "video_chapters": chapters_and_captions,
    }

    return final_output


# Example usage
final_output = get_video_details(video_id, youtube_api_key)
print(final_output)
````
